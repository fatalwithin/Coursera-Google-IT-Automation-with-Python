```python
"""
Task 2 - fixing website reviews
"""

import os
import requests
import json

root_dir = "/data/feedback"
external_ip = "127.0.0.1"

def get_txt_files(dir):
    """
    Lists all txt files under specified directory

    Parameters:
    dir(str): path to specified directory

    Returns:
    []: a list of txt files inside specified directory
    """
    result = []
    try:
        dirs = os.listdir(path=root_dir)
        for file in dirs:
            if str(file).endswith("txt"):
                result.append(file)
    except OSError:
        print("No such file or directory: " + root_dir)
    return result

def deserialize_feedback_content(files):
    """
    Traverses over each file and adds a record to dictionary with title, name, date and feedback as keys

    Parameters:
    files[]: a list of files to process

    Result:
    []: a list of dictionaries made of input list of files
    """
    result = []

    for file in files:
        try:
            with open(file, mode='r') as file_in_process:
                temp = file_in_process.splitlines()
                review = dict(title=temp[0], name=temp[1], date=temp[2], feedback=temp[3])
                result.append(review)
        except OSError:
            print("Cannot open file: " + file)
    return result

def post_review(review, corpweb_ip):
    """
    Posts a review to the company's website. Makes a POST request to http://<corpweb-external-ip>/feedback

    Parameters:
    reviews[]: a list of deserialized reviews as dictionaries
    corpweb_ip(int): an IP address of corporate website

    Result:
    - posted reviews;
    - for each successfully posted review a response status is printed

    """

    post_address = "http://" + corpweb_ip + "/feedback"

    post_data = review

    response = requests.post(post_address, data=post_data)
    if not response.ok:
        raise Exception("GET failed with status code {}".format(response.status_code))
    else:
        print(review["title"] + " status is " + response.status_code)


if __name__ == "__main__":
    file_list = get_txt_files(root_dir)

    deserialized_data_list = deserialize_feedback_content(file_list)

    for item in deserialized_data_list:
        post_review(item, corpweb_ip)

```
